```python
import pandas as pd
```

```python
df = pd.read_csv('Training_Data_Set.csv')
df
```

```python
df.info()
```

```python
df.isna().values.any()
```

```python
df.isna().sum(axis=0)
```

```python
df.fillna({"Location":df['Location'].mode()[0], 
           "Distance": df['Distance '].median(), 
           "Manufacture": df['manufacture_year'].median(),
          "Engine_power":df['engine_power'].median(),
          "body_type":df['body_type'].mode()[0],
           "door":df['door_count'].median(),
           "seat":df['seat_count'].median()},
          inplace=True)
df
```

```python
df.isna().sum(axis=0)
```

```python
df.fillna(method='ffill', inplace=True)
df
```
